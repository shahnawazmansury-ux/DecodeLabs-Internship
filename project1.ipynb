```python
import pandas as pd

```

```python
file_path =  r"C:\Users\IIT\Desktop\pandas\Dataset for Data Analytics.xlsx"

df = pd.read_excel(file_path)

print("=" * 50)
print("DATASET LOADED SUCCESSFULLY")
print("=" * 50)
```

```python
print("\nDataset Shape:")
print(df.shape)

print("\nColumn Names:")
print(df.columns.tolist())

print("\nDataset Information:")
print(df.info())

print("\nFirst 5 Records:")
print(df.head())
```

```python
print("\n" + "=" * 50)
print("MISSING VALUES")
print("=" * 50)

missing_values = df.isnull().sum()

print(missing_values)
```

```python
print("\nHandling Missing Values...")

# CouponCode column

if "CouponCode" in df.columns:
    df["CouponCode"] = df["CouponCode"].fillna("No Coupon")

print("Missing values handled successfully.")
```

```python
print("\nRemaining Missing Values:")

print(df.isnull().sum())
```

```python
print("\n" + "=" * 50)
print("DUPLICATE RECORDS")
print("=" * 50)

duplicate_rows = df.duplicated().sum()

print(f"Total Duplicate Rows Found: {duplicate_rows}")
```

```python
df = df.drop_duplicates()

print("Duplicates Removed Successfully.")

print("Current Shape:", df.shape)
```

```python
if "OrderID" in df.columns:

    duplicate_ids = df["OrderID"].duplicated().sum()

    print(f"\nDuplicate Order IDs: {duplicate_ids}")
```

```python
print("\n" + "=" * 50)
print("DATE FORMAT VALIDATION")
print("=" * 50)

if "Date" in df.columns:

    df["Date"] = pd.to_datetime(
        df["Date"],
        errors="coerce"
    )

    invalid_dates = df["Date"].isnull().sum()

    print(f"Invalid Dates Found: {invalid_dates}")
```

```python
print("\n" + "=" * 50)
print("DATA TYPES")
print("=" * 50)

print(df.dtypes)
```

```python
numeric_columns = [
    "Quantity",
    "UnitPrice",
    "ItemsInCart",
    "TotalPrice"
]

for column in numeric_columns:

    if column in df.columns:

        df[column] = pd.to_numeric(
            df[column],
            errors="coerce"
        )

print("\nNumeric columns validated.")
```

```python
text_columns = [
    "Product",
    "ShippingAddress",
    "PaymentMethod",
    "OrderStatus",
    "CouponCode",
    "ReferralSource"
]

for column in text_columns:

    if column in df.columns:

        df[column] = (
            df[column]
            .astype(str)
            .str.strip()
        )

print("Text columns cleaned.")
```

```python
print("\n" + "=" * 50)
print("FINAL QUALITY CHECK")
print("=" * 50)

print("\nFinal Shape:")
print(df.shape)

print("\nRemaining Missing Values:")
print(df.isnull().sum())

print("\nDuplicate Rows:")
print(df.duplicated().sum())
```
